```python
import pandas as pd
```

```python
countries = pd.read_csv('/content/countries_regions.csv')
countries
```

```python
countries.columns = ['geo','Country','World bank region']
```

```python
global_development = pd.read_csv('/content/global_development.csv')
global_development 
```

```python
header = list(global_development.columns)
global_development = global_development[header[:12]]
```

```python
global_development
```

```python
newDf = global_development.merge(countries, on = 'Country', how = 'left')
newDf
```

```python
newDf.dropna(subset=['geo'], inplace=True)
newDf
```

```python
gb = newDf.groupby('Country')
df = [gb.get_group(x) for x in gb.groups]
```

```python
for f in df:
  f.to_csv('/content/drive/MyDrive/DV_CSV/' + str(f.iloc[0,0]) + '.csv', index=None)
  # f.to_csv(str(f['Country'][0])+'.csv')
```
